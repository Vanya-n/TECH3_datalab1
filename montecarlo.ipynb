## Monte Carlo simulation of queue
Run the code below. Add a line for calculating the standard deviation of waiting time. 
Adjust the arrival\_rate to 0.8 and 0.95. What is the expected wainting time? What about the standard deviation?

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Parameters
# -------------------------
np.random.seed(42)

n_customers = 10000
# ----------------------------------
# -- Play with these arguments -----
arrival_rate = .5   # lambda (customers per unit time)
service_rate = 1.0   # mu (customers served per unit time)
# ----------------------------------
# Utilization
rho = arrival_rate / service_rate
print(f"Utilization (rho): {rho:.2f}")

# -------------------------
# Simulate interarrival and service times
# -------------------------
interarrival_times = np.random.exponential(1 / arrival_rate, n_customers)
service_times = np.random.exponential(1 / service_rate, n_customers)

# Arrival times
arrival_times = np.cumsum(interarrival_times)

# -------------------------
# Queue dynamics
# -------------------------
start_service_times = np.zeros(n_customers)
finish_service_times = np.zeros(n_customers)
waiting_times = np.zeros(n_customers)

for i in range(n_customers):
    if i == 0:
        start_service_times[i] = arrival_times[i]
    else:
        start_service_times[i] = max(arrival_times[i], finish_service_times[i-1])

    finish_service_times[i] = start_service_times[i] + service_times[i]
    waiting_times[i] = start_service_times[i] - arrival_times[i]

# -------------------------
# Results
# -------------------------
print(f"Average waiting time: {np.mean(waiting_times):.4f}")
print(f"Average time in system: {np.mean(waiting_times + service_times):.4f}")
print(f"Standard deviation of waiting time: {np.std(waiting_times):.4f}")

# -------------------------
# Plot waiting time distribution
# -------------------------
plt.hist(waiting_times, bins=50)
plt.title("Distribution of Waiting Times")
plt.xlabel("Waiting Time")
plt.ylabel("Frequency")
plt.show()
```

# Bootstrap standard error of success probability estimate
Bootstrap the variance of the estimate for the probability of guessing the outcome of the dice correctly. Use np.random.choice for sampling from the data. 

```python
import pandas as pd
np.random.seed(123)
url = ("https://raw.githubusercontent.com/holleland/TECH3/refs/heads/main/data/ekstrom2025.csv")
data = pd.read_csv(url, sep = ";")

# Use only the first 10:
n = 10 # n = 800 is the full dataset
sample = data["correct_guess"].head(n)
# Estimate probability of correct guess: 
phat = sample.mean() 
print("phat:", phat)

# Bootstrap: 
B = 5000 # Bootstrap samples
boot_phat =  np.zeros(B)
# bootstrap step:
for b in range(B):
    # Generate bootstrap sample
    boot = np.random.choice(sample, size=n, replace=True) # Fill in here 
    # Calculate estimator:
    boot_phat[b] = boot.mean() # FILL IN 
# Print summary statistics of bootstrap samples: 
print(f"Bootstrap Mean: {boot_phat.mean():.4f}")
print(f"Bootstrap Var: {boot_phat.var():.4f}")
print(f"Formula Var: {(phat*(1-phat)/n):.4f}")
```
