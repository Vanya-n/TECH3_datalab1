```python
import pandas as pd
brann = pd.read_csv(
"https://raw.githubusercontent.com/holleland/TECH3/refs/heads/main/data/Brann_Recycling.csv"
)
brann.head()
```

## Question1

```python
brann.groupby('Opponent')['SortedCup'].sum()
```

```python
a = brann.groupby('Opponent')['SortedCup'].sum()
```

## Question2

```python
b = brann.groupby('Opponent')['SortedCup'].count()
```

```python
c = pd.merge(a, b, on='Opponent')
c.rename(columns={'SortedCup_x': 'TotalSortedCup', 'SortedCup_y': 'CountSortedCup'}, inplace=True)
c['AverageSortedCup'] = c['TotalSortedCup'] / c['CountSortedCup']*100
c
```

## Question3

```python
c['AverageSortedCup'].describe()
```

```python
c['AverageSortedCup'].mean()
```

```python
c['AverageSortedCup'].median()
```

```python
c['AverageSortedCup'].std()
```

```python
c['AverageSortedCup'].quantile(0.25)
```

```python
c['AverageSortedCup'].quantile(0.75)
```

## Question4

```python
import matplotlib.pyplot as plt
plt.bar(c.index, c['AverageSortedCup'], color='red')
plt.xlabel('Opponent')
plt.ylabel('Average Sorted Cup Percentage')
plt.title('Average Sorted Cup Percentage by Opponent')
plt.xticks(rotation=45)
```
