## 1e)

```python
# (i) from the sample 
B=5000
boot_alpha = np.zeros(B)
for b in range(B):
    boot_sample = np.random.choice(x, size=len(x), replace=True)
    boot_alpha[b] = -len(boot_sample) / np.sum(np.log(boot_sample))
print("Bootstrap alpha estimates:", boot_alpha)
lower, upper = np.percentile(boot_alpha, [5, 95])
print(f"90% confidence interval for alpha: [{lower:.4f}, {upper:.4f}]")
print("Variance: ", boot_alpha.var())
```

```python
# (ii) Parametric bootstrap
from scipy.stats import beta
pboot_alpha = np.zeros(B)
for b in range(B):
    boot_sample = beta.rvs(a=alpha_hat, b=1, size=10)
    pboot_alpha[b] = -len(boot_sample) / np.sum(np.log(boot_sample))
lower, upper = np.percentile(pboot_alpha, [5, 95])
print(f"90% confidence interval for alpha: [{lower:.4f}, {upper:.4f}]")
print("Variance: ", pboot_alpha.var())
```

```python
# (iii) Normal approximation
from scipy.stats import norm
clt_lower, clt_upper= norm.ppf((0.05,0.95), loc=alpha_hat, scale=boot_alpha.std())
print(f"90% confidence interval for alpha: [{clt_lower:.4f}, {clt_upper:.4f}]")
clt_lower, clt_upper= norm.ppf((0.05,0.95), loc=alpha_hat, scale=pboot_alpha.std())
print(f"90% confidence interval for alpha pboot: [{clt_lower:.4f}, {clt_upper:.4f}]")
```

```python
#c)
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
p_win = 18 / 37
initial_balance = 100
max_spins = 5000  # safety cap to avoid infinite loops

# --- Initialization ---
balance = initial_balance
bet = 1
balances = [balance]

for spin in range(max_spins):

    # Check if you can afford the bet
    if bet > balance:
        print("Bankrupt!")
        break

    # Place bet
    balance -= bet

    # Simulate outcome
    win = np.random.binomial(size=1, p=p_win, n = 1) == 1

    if win:
        # Win: receive 2*bet (stake + winnings)
        balance += 2 * bet
        bet = 1  # reset bet after win
    else:
        # Loss: double the bet
        bet *= 2

    balances.append(balance)
else:
    print("Maximum spins reached")
print("Bankrupt after round ", spin)
```
