```python
import pandas as pd

x= pd.DataFrame([4.2,4.2,4.4,4.7,5.2,5.3,5.5,5.6,5.7,6.5,6.6])
x.median()
```

```python
x.describe()
#IQR = Q3 - Q1= 5.6 - 4.4 = 1.2
#Lower bound = Q1 - 1.5*IQR = 4.4 - 1.5*1.2 = 2.6
#Upper bound = Q3 + 1.5*IQR = 5.6 + 1.5*1.2 = 7.4
#No outliers
```
